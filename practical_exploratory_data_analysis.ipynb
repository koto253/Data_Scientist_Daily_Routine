```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('https://raw.githubusercontent.com/franklin-univ-data-science/data/master/olympics.csv')
```

```python
# check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Summary Statistics
summary_stats = df.describe()
summary_stats
```

```python
# Data Visualization
# Histogram of ages

plt.figure(figsize=(8, 6))
sns.histplot(df['Edition'], bins=5, kde=True)
plt.title('Medals Distribution by Year')
plt.xlabel('Medal')
plt.ylabel('Distribution')
plt.show()
```

```python
# Correlation Analysis
# Explore the correlation between different variables

correlation_matrix = df.corr()
correlation_matrix
```
